```python
import numpy as np
import skimage as ski
```

```python
camera = ski.data.camera()
ski.io.imshow(camera)
```

# Thresholding

```python
intnensityValue = 180
camera[camera < intnensityValue] = 0
camera[camera >= intnensityValue] = 255
```

```python
ski.io.imshow(camera)
```

# Inverse

S = L - 1 - r

L = 8-bit => 256
r = Pixel Value

```python
inversecamera = 256 - 1 - camera
ski.io.imshow(inversecamera)
```

# Log

s = log(r+1)

```python
logImage = np.log1p(camera)
ski.io.imshow(logImage/np.max(logImage))
```

# Constrast Streching

$$ S = (I-min) \times \frac{new Max - new Min}{Max - Min} + new Min $$

10-bit image

$$ new Max = 1024, new Min= 0$$

```python
# New Image Intensity Range of 8-bit image
newMax = 255
newMin = 0

# Find max and min of the image
Max = np.max(camera)
Min = np.min(camera)  

s = (camera - Min) / (Max - Min) * (newMax - newMin) + newMin

ski.io.imshow(s.astype(np.uint8))
```
