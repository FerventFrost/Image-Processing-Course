## Read Gray image

```python
grayImage = ski.io.imread("kP0u2.png", as_gray=True)
ski.io.imshow(grayImage)
```

```python
grayImage.ndim
```

# Increase image intesity

```python
Inc_grayImage = grayImage.copy()
Inc_grayImage.dtype
```

```python
Inc_grayImage = (Inc_grayImage * 255).astype('uint8')
ski.io.imshow(Inc_grayImage)
```

```python
# increase intesity by 20
Inc_grayImage += 20
ski.io.imshow(Inc_grayImage)
```

```python
NoiseImage = np.random.randint(0,255,(Inc_grayImage.shape), dtype='uint8')
ski.io.imshow(NoiseImage)
```

### Add Noise to image

```python
# Limit effect of noise to 20% of the image
limitEffect = 0.2
noisedImage = Inc_grayImage + limitEffect * NoiseImage
```

```python
ski.io.imshow(noisedImage.astype('uint8'))
```

## Access Array

Array[Rows, Columns, Color]

Array [ Rows, Columns, Color]


Array[200] => Single Row

Array[200:300] => Multiple Rows


Array[:, 150] => Single Column

Array[:, 150:250] => mutiple Column

Array[200:300, 150:250] => Union between Rows and Columns

```python
blackBarImage = image.copy()
blackBarImage[0:100] = 0
ski.io.imshow(blackBarImage)
```

```python
blackBarImage[:,100:200] = 0
ski.io.imshow(blackBarImage)
```

```python
splitBlackBar = image.copy()
splitBlackBar[:200,200:300] = 0
splitBlackBar[300:,200:300] = 0
ski.io.imshow(splitBlackBar)
```

## Split Color

```python
SplitColor = image.copy()
readChannel = SplitColor.copy()
greenChannel = SplitColor.copy()
blueChannel = SplitColor.copy()
```

```python
readChannel[:,:,1] = 0
readChannel[:,:,2] = 0

greenChannel[:,:,0] = 0
greenChannel[:,:,2] = 0

blueChannel[:,:,0] = 0
blueChannel[:,:,1] = 0
```

```python
ski.io.imshow(readChannel)
ski.io.show()
ski.io.imshow(greenChannel)
ski.io.show()
ski.io.imshow(blueChannel)
ski.io.show()
```
