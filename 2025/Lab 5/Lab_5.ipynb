```python
import matplotlib.pyplot as plt
import skimage as ski
import numpy as np
```

```python
x = [1,2,3,4,5]
y = [1,4,9,16,25]
z = [1,8,27,64,125]
plt.plot(x,y, color='red', linestyle="--", marker='o', label="Red Color")
plt.plot(x,z,color="green", label="Green Color")
plt.legend()

plt.xlabel("Data")
plt.ylabel("Y=X^2")

plt.show()
```

```python
plt.hist([10,10,9,20,5,5,5,1], bins=5)
# 20 / 5 = 4
```
