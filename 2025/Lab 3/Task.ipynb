# Euclidean
$$ Radius = \sqrt{(X - X_{Midpoint}) ^ 2 + (Y - Y_{Midpoint})^2} $$

## Circle
$$ Radius > \sqrt{(X - X_{Midpoint}) ^ 2 + (Y - Y_{Midpoint})^2} $$

```python
Circle = np.zeros(image.shape, dtype=np.uint8)
midpoint = (150, 250)
```

```python
for i in range(100,200):
  for j in range(200,300):
    if (i-midpoint[0])**2 + (j-midpoint[1])**2 < 50**2:
      Circle[i,j] = 1
```

```python
ski.io.imshow(Circle * 255)
```

```python
ski.io.imshow(image * Circle)
```
